```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
import numpy as np
```

```python
train = pd.read_csv('../data/v3/training_seg2.csv')
```

```python
train.head()
```

```python
train.describe()
```

```python
train['application_date'] = pd.to_datetime(train['application_date'])
```

```python
df = train.set_index('application_date')
```

```python
y = df['case_count'].resample('D').mean()
```

```python
y.fillna(0, inplace=True)
```

```python
y.plot(figsize=(15, 6))
```

```python
decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
plt.show()
```

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
```

```python
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(16, 8))
```

```python
pred = results.get_prediction(start=pd.to_datetime('2019-01-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = y['2017':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Case_Counts')
plt.legend()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['2017-01-01':]
mse = ((y_forecasted - y_truth) ** 2).mean()
mape = (abs(y_forecasted - y_truth)/y_truth)*(100/len(y_forecasted))
```

```python
pred_uc = results.get_forecast(steps=100)
pred_ci = pred_uc.conf_int()
ax = y.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Case_Counts')
plt.legend()
plt.show()
```

```python
forecasted = np.round(pred_uc.predicted_mean)
forecasted = pd.DataFrame(forecasted,columns=['forecast'])
```

```python
test = pd.read_csv('../data/v3/test_seg1.csv')
```

```python
test['application_date'] = pd.to_datetime(test['application_date'])
```
