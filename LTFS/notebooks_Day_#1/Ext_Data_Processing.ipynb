```python
import pandas as pd
```

```python
df17 = pd.read_csv('2017_holidays.csv')
df18 = pd.read_csv('2018_holidays.csv')
df19 = pd.read_csv('2019_holidays.csv')
```

```python
df17['year'] = pd.DatetimeIndex(df17['Date']).year
df17['day'] = pd.DatetimeIndex(df17['Date']).day
df17['month'] = pd.DatetimeIndex(df17['Date']).month
df17['year'] = 2017
```

```python
df18['year'] = pd.DatetimeIndex(df18['Date']).year
df18['day'] = pd.DatetimeIndex(df18['Date']).day
df18['month'] = pd.DatetimeIndex(df18['Date']).month
df18['year'] = 2018
```

```python
df19['year'] = pd.DatetimeIndex(df19['Date']).year
df19['day'] = pd.DatetimeIndex(df19['Date']).day
df19['month'] = pd.DatetimeIndex(df19['Date']).month
df19['year'] = 2019
```

```python
df7 = df17.drop(columns=['Date'])
df8 = df18.drop(columns=['Date'])
df9 = df19.drop(columns=['Date'])
```

```python
df = pd.concat([df7, df8, df9], ignore_index=True, sort=False)
df
```

```python
df.to_csv('list_of_hol.csv', index=False)
```
