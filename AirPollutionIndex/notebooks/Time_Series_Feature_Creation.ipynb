```python
total['day'] = total['date_time'].apply(lambda x: x.day)
total['month'] = total['date_time'].apply(lambda x: x.month)
total['year'] = total['date_time'].apply(lambda x: x.year)
total['dow'] = total['date_time'].apply(lambda x: x.dayofweek)
total['doy'] = total['date_time'].apply(lambda x: x.dayofyear)
total['hour'] = total['date_time'].apply(lambda x: x.hour)
```

```python
train.head()
```

```python
train = total[~total['air_pollution_index'].isna()].reset_index(drop=True)
test = total[total['air_pollution_index'].isna()].reset_index(drop=True)
```

```python
train.shape, test.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
la = LabelEncoder()
la2 = LabelEncoder()
```

```python
train['enc_hol'] = la.fit_transform(train['is_holiday'])
train['enc_weat'] = la2.fit_transform(train['weather_type'])
```

```python
train.head()
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
norm = Normalizer()
```

```python
X=norm.fit_transform(train[['humidity', 'wind_speed', 'wind_direction', 'visibility_in_miles', 
        'temperature', 'rain_p_h', 'snow_p_h', 'clouds_all', 
        'traffic_volume', 'day', 'month', 'year', 'dow', 'doy', 'hour',
        'enc_hol', 'enc_weat']])
y = train['air_pollution_index']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
import lightgbm
import catboost
```

```python
lgt = lightgbm.LGBMRegressor()
cat = catboost.CatBoostRegressor()
rf = RandomForestRegressor()
```

```python
lgt.fit(X_train, y_train)
rf.fit(X_train, y_train)
cat.fit(X_train, y_train)
```

```python
lgt_pred = lgt.predict(X_test)
cat_pred = cat.predict(X_test)
rf_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
def score(y_true, y_pred):
    return max(0, (100 - np.sqrt(mean_absolute_error(y_true, y_pred))))
```

```python
score(y_test, lgt_pred), score(y_test, cat_pred), score(y_test, rf_pred)
```

# Stacking

```python
from mlxtend.regressor import StackingCVRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from scipy.stats import randint
from scipy.stats import uniform


rf = RandomForestRegressor()
cat = catboost.CatBoostRegressor()
cat2 = lightgbm.LGBMRegressor()
ada = AdaBoostRegressor()
bag = BaggingRegressor()
# lr = LinearRegression()
lgt = lightgbm.LGBMRegressor()

stack = StackingCVRegressor(regressors=(rf, cat, lgt, ada, bag),
                            meta_regressor=cat2, 
                            random_state=42,
                            use_features_in_secondary=True)
```

```python
stack.fit(X_train, y_train)
```

```python
stack_pred = stack.predict(X_test)
```

```python
score(y_test, stack_pred)
```
