# Boxplot

```python
import math
```

```python
def choose_subplot_dimensions(k):
    if k < 4:
        return k, 1
    elif k < 11:
        return math.ceil(k/2), 2
    else:
        # I've chosen to have a maximum of 3 columns
        return math.ceil(k/3), 3


def generate_subplots(k, row_wise=False):
    nrow, ncol = choose_subplot_dimensions(k)
    # Choose your share X and share Y parameters as you wish:
    figure, axes = plt.subplots(nrow, ncol,
                                sharex=False,
                                sharey=False,
                                figsize=(6, 15))

    # Check if it's an array. If there's only one plot, it's just an Axes obj
    if not isinstance(axes, np.ndarray):
        return figure, [axes]
    else:
        # Choose the traversal you'd like: 'F' is col-wise, 'C' is row-wise
        axes = axes.flatten(order=('C' if row_wise else 'F'))

        # Delete any unused axes from the figure, so that they don't show
        # blank x- and y-axis lines
        for idx, ax in enumerate(axes[k:]):
            figure.delaxes(ax)

            # Turn ticks on for the last ax in each column, wherever it lands
            idx_to_turn_on_ticks = idx + k - ncol if row_wise else idx + k - 1
            for tk in axes[idx_to_turn_on_ticks].get_xticklabels():
                tk.set_visible(True)

        axes = axes[:k]
        return figure, axes
```

We analyse the boxplot below, which indicates the outliers present above and below the 3*IQR

```python
# Red Wine
attributes = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
     'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
     'pH', 'sulphates', 'alcohol', 'quality']

figure, axes = generate_subplots(len(attributes), row_wise=True)
for parameter, ax in zip(attributes, axes):
    sns.boxplot(x=parameter, data=red,  orient='v',ax=ax )
#     ax.set_title(label=parameter)

plt.tight_layout()
plt.show()
```

The points indicate data above/below (1.5*IQR)

```python
attributes = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
     'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
     'pH', 'sulphates', 'alcohol', 'quality']

figure, axes = generate_subplots(len(attributes), row_wise=True)
for parameter, ax in zip(attributes, axes):
    sns.boxplot(x=parameter, data=white,  orient='v',ax=ax )
#     ax.set_title(label=parameter)

plt.tight_layout()
plt.show()
```

# Kernel Density Estimation

```python
attributes = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
     'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
     'pH', 'sulphates', 'alcohol', 'quality']

figure, axes = generate_subplots(len(attributes), row_wise=True)
for parameter, ax in zip(attributes, axes):
    sns.distplot(red[parameter], ax=ax, kde=True )
#     ax.set_title(label=parameter)

plt.tight_layout()
plt.show()
```

```python
attributes = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
     'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
     'pH', 'sulphates', 'alcohol', 'quality']

figure, axes = generate_subplots(len(attributes), row_wise=True)
for parameter, ax in zip(attributes, axes):
    sns.distplot(white[parameter], ax=ax, kde=True )
#     ax.set_title(label=parameter)

plt.tight_layout()
plt.show()
```
