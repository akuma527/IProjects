# Identifying Outliers in num_orders based on groups

```python
len(train[['num_orders', 'group']]['group'].unique())
```

```python
plt.boxplot(train[train['group']=='55_1885']['num_orders'])
```

```python
def _get_high_low(gr):
    k = train[train['group']==gr].reset_index(drop=True)
    IQR = (k[['num_orders']].describe().loc['75%'][0]-k[['num_orders']].describe().loc['25%'][0])
    low_out = k[['num_orders']].describe().loc['25%'][0] - 1.5*IQR
    high_out = k[['num_orders']].describe().loc['75%'][0] + 1.5*IQR
    return k, high_out, low_out
```

```python
def outlier(df, high_out, low_out):
    if (df>high_out) or (df<low_out):
        return 1
    else:
        return 0
```

```python
groups = list(train.group.unique())
```

```python
a = []
for gr in groups:
    k, high_out, low_out = _get_high_low(gr)
    k['outlier'] = k['num_orders'].apply(outlier,args=(high_out, low_out,))
    a.append(k)
```

```python
df = pd.concat(a, ignore_index=True, sort=False)
```

```python
del a
```

```python
len(df[df['outlier']==1])/len(df), len(df[df['outlier']==1]), len(df)
```

```python
df = df[df['outlier']==0].reset_index(drop=True)
```

# Checking emailer_for_promotion & homepage_featured

```python
df['emailer_for_promotion'].unique(), df['homepage_featured'].unique()
```

```python
observed = pd.crosstab(df['homepage_featured'], df['emailer_for_promotion'], 
                       rownames=['homepage_featured'], colnames=['emailer_for_promotion'])
```

```python
from scipy.stats.contingency import margins, reduce, chi2_contingency
```

```python
chi2, p, dof, expected = chi2_contingency(observed)
```

Hence we reject the Null hypothesis, and conclude that the features are dependent
