```python
def encode_MAS(df):
    if str(df) == 'BrkFace':
        return 0
    elif str(df) == 'None':
        return 1
    elif str(df) == 'Stone':
        return 2
    elif str(df) == 'BrkCmn':
        return 3
    else:
        return None    
```

```python
train['Mas_Type'] = train['MasVnrType'].apply(encode_MAS)
```
